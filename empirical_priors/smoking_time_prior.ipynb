```python
def smoking_time(contingent_ema, random_ema, window):
    '''
    Takes the contingent and random emas as signals for the smoking events and
    output the event timestamps in sequential format grouped by quit date
    '''
    
    smoking_events = {}
    
    # add timestamps from the contingent ema
    for index, row in contingent_ema.iterrows():
        if row['status'] == "MISSED":
            continue
        if row['participant_id'] not in smoking_events:
            smoking_events[row['participant_id']] = {}
            
        try:
            time = datetime.datetime.strptime(row['date'], '%m/%d/%y %H:%M')
        except:
            time = datetime.datetime.strptime(row['date'], '%Y-%m-%d %H:%M:%S')
        date = (time.year, time.month, time.day) # granularize time to day
        
        if date not in smoking_events[row['participant_id']]:
            smoking_events[row['participant_id']][date] = [] # store smoking timestamps by quit date
            
        response_time = int(row['timestamp'] / 1000) # ignore seconds
        smoking_time = response_time - contingent_keys[row['when_smoke']] * 60 # compute smoking time
        smoking_events[row['participant_id']][date].append(smoking_time)
    
    # add timestamps from the random ema
    for index, row in random_ema.iterrows():
        if row['status'] != "COMPLETED" or row['smoke'] == 'No':
            continue
        if row['participant_id'] not in smoking_events:
            smoking_events[row['participant_id']] = {}
            
        try:
            time = datetime.datetime.strptime(row['date'], '%m/%d/%y %H:%M')
        except:
            time = datetime.datetime.strptime(row['date'], '%Y-%m-%d %H:%M:%S')
        date = (time.year, time.month, time.day) # granularize time to day
        
        if date not in smoking_events[row['participant_id']]:
            smoking_events[row['participant_id']][date] = [] # store smoking timestamps by quit date
            
        response_time = int(row['timestamp'] / 1000) # ignore seconds
        smoking_time = response_time - random_keys[row['when_smoke']] * 60 # compute smoking time
        smoking_events[row['participant_id']][date].append(smoking_time)
        
    return smoking_events
```
