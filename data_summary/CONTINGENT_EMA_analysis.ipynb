```python
import pandas as pd
import numpy as np
```

```python
# read data
phone_data = pd.read_csv('eventcontingent-ema.csv')
phone_data = phone_data.drop(['offset'], axis=1)
cloud_data = pd.read_csv('eventcontingent-ema-backup.csv')
```

```python
# check data types
print(phone_data.dtypes)
print("")
print(cloud_data.dtypes)
```

```python
# convert data types to be consistent
cloud_data['urge'] = cloud_data['urge'].astype(float)
cloud_data['cheerful'] = cloud_data['cheerful'].astype(float)
cloud_data['happy'] = cloud_data['happy'].astype(float)
cloud_data['angry'] = cloud_data['angry'].astype(float)
cloud_data['stress'] = cloud_data['stress'].astype(float)
cloud_data['sad'] = cloud_data['sad'].astype(float)
cloud_data['access'] = cloud_data['access'].astype(float)
cloud_data['see_or_smell'] = cloud_data['see_or_smell'].astype(float)
cloud_data['smoking_location'] = cloud_data['urge'].astype(float)

print(phone_data.dtypes)
print(cloud_data.dtypes)
```

```python
print(phone_data.shape)
phone_data['participant_id'].value_counts().sort_index()
```

```python
print(cloud_data.shape)
cloud_data['participant_id'].value_counts().sort_index()
```

```python
# take out participants in Phone Data that's not in Cloud Data
phone_data_backup = phone_data[~phone_data.participant_id.isin([201,203,206,210,221,229])]
```

```python
print(phone_data_backup.shape)
phone_data_backup['participant_id'].value_counts().sort_index()
```

```python
s = set()
d = {}

lst2 = []
for index, row in cloud_data.iterrows():
    participant_id = row['participant_id']
    hour = row['hour']
    minute = row['minute']
    day_of_week = row['day_of_week']
    valid_key = (participant_id, hour, minute, day_of_week)
    s.add(valid_key)
    d[valid_key] = index

for index, row in phone_data.iterrows():
    participant_id = row['participant_id']
    hour = row['hour']
    minute = row['minute']
    day_of_week = row['day_of_week']
    valid_key = (participant_id, hour, minute, day_of_week)
    if valid_key in s:
        s.remove(valid_key)
    else:
        lst2.append(index)

lst = []
for i in s:
    lst.append(d[i])

lst.sort()
difference = cloud_data.ix[lst] 
print(difference.shape)
difference # rows that are only in cloud but not phone
```
