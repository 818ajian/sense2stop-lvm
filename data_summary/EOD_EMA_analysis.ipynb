```python
import pandas as pd
import numpy as np
```

```python
# read data
phone_data = pd.read_csv('eod-ema.csv')
phone_data = phone_data.drop(['offset'], axis=1)
cloud_data = pd.read_csv('eod-ema-backup.csv')
```

```python
print('End of Day EMA Phone Data Summary:')
print(phone_data.describe())
```

```python
print(phone_data.shape)
phone_data['participant_id'].value_counts().sort_index()
```

```python
print('End of Day EMA Cloud Backup Data Summary:')
print(cloud_data.describe())
```

```python
print(cloud_data.shape)
cloud_data['participant_id'].value_counts().sort_index()
```

```python
# take out participants in Phone Data that's not in Cloud Data
phone_data_backup = phone_data[~phone_data.participant_id.isin([201,203,206,210,221,229])]
print('Phone Data Summary')
print(phone_data_backup.describe())
```

```python
print(phone_data_backup.shape)
phone_data_backup['participant_id'].value_counts().sort_index()
```

```python
common_rows = pd.merge(cloud_data, phone_data_backup, how='inner')
print('Common Rows Summary between Phone and Cloud')
print(common_rows.describe())
```

```python
s = set()
d = {}

lst2 = []
for index, row in cloud_data.iterrows():
    participant_id = row['participant_id']
    hour = row['hour']
    minute = row['minute']
    day_of_week = row['day_of_week']
    valid_key = (participant_id, hour, minute, day_of_week)
    s.add(valid_key)
    d[valid_key] = index

for index, row in phone_data.iterrows():
    participant_id = row['participant_id']
    hour = row['hour']
    minute = row['minute']
    day_of_week = row['day_of_week']
    valid_key = (participant_id, hour, minute, day_of_week)
    if valid_key in s:
        s.remove(valid_key)
    else:
        lst2.append(index)

lst = []
for i in s:
    lst.append(d[i])

lst.sort()
difference = cloud_data.ix[lst] 
print(difference.shape)
difference # rows that are only in cloud but not phone
```

```python
difference = phone_data.ix[lst2]
print(difference.shape)
difference # rows in phone but not cloud
```
